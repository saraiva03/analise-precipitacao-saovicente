```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster

# =================================================
# 1. criando dataframe com coordenadas das estações
# =================================================

dados = [
    {"nome_posto": "HUMAITA", "latitude": -23.954444, "longitude": -46.450000},
    {"nome_posto": "GASPAR RICARDO", "latitude": -23.966667, "longitude": -46.500000},
    {"nome_posto": "GASPAR RICARDO (EFS)", "latitude": -23.966667, "longitude": -46.500000},
    {"nome_posto": "SALES DA CRUZ", "latitude": -23.916667, "longitude": -46.466667},
    {"nome_posto": "PAI MATIAS (EFS)", "latitude": -23.950000, "longitude": -46.533333},
    {"nome_posto": "ENGENHEIRO FERRAZ (EFS)", "latitude": -23.983333, "longitude": -46.616667},
    {"nome_posto": "SAO VICENTE", "latitude": -23.960278, "longitude": -46.364722}
]

df = pd.DataFrame(dados)

# ==========================================================================
# 2. criando mapa com a topologia de são vicente para plottar as coordenadas
# ==========================================================================

mapa = folium.Map(location=[-23.96, -46.45], zoom_start=12)

folium.TileLayer(
    'Stamen Terrain',
    name="Relevo / Altimetria"
).add_to(mapa)

# ====================================================================
# 3. adicionan marker cluster para deixar o nome das estações visíveis 
#    nas suas respectivas coordenadas     
# ====================================================================

cluster = MarkerCluster().add_to(mapa)

for _, row in df.iterrows():
    folium.Marker(
        location=[row["latitude"], row["longitude"]],
        popup=f"Posto: {row['nome_posto']}",
        tooltip=row["nome_posto"]
    ).add_to(cluster)

# =======================
# 4. exibindo o mapa
# =======================
mapa
```
